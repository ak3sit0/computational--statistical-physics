```julia
using DifferentialEquations, Plots
```

```julia
function kuramoto_model!(dθ, θ, parameters, t)
    N, K, ω = parameters
    s_sin = sum(sin.(θ))
    s_cos = sum(cos.(θ))
    @inbounds dθ .= ω .+ (K / N) .* (s_sin .* cos.(θ) .- s_cos .* sin.(θ))
end
```

```julia
initial_conditions = [0.0, π/2, π, 3π/2]
parameters = (4, 1.0, [1.0, 1.5, 0.5, 2.0])
simulation_time = (0.0, 100.0)

kuramoto_problem = ODEProblem(kuramoto_model!, initial_conditions, simulation_time, parameters)
kuramoto_solution = solve(kuramoto_problem, Tsit5(), saveat=0.1)
```

```julia
u, t = kuramoto_solution.u, kuramoto_solution.t # Extraer solución y tiempos
theta_matrix = hcat(u...)  #  
```

```julia
theta_matrix = hcat(u...)  # Mostrar las primeras 5 columnas de la solución
```

```julia
plot(t, hcat(u...)', xlabel="Time", ylabel="Phase", title="Kuramoto Model Simulation", legend=false)
```
